# US Country Level Presidential Result

### 2016 US County Level Presidential

```python
import pandas as pd 
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
```

```python
df1 = pd.read_csv("2008_2016_US_County_Level_Presidential_Results.csv") 
df2 = pd.read_csv("2020_US_County_Level_Presidential_Results.csv") 
```

```python
df1.head()
```

```python
us_data_2008 = df1
data_2008 = us_data_2008.values
```

```python
data_2008
```

```python
data_2008.shape
```

```python
df1_0, df1_1 = data_2008.shape
x1, y1 = np.linspace(0, 10, df1_0), np.linspace(0, 10, df1_1)
fig1 = go.Figure(data=[go.Surface(z=data_2008, x=x1, y=y1)])
fig1.update_layout(title='2008_2016_US_County_Level_Presidential_Results',width=750, height=750)

fig1.update_layout(scene = dict(
                    xaxis_title='ROWS',
                    yaxis_title='COLLUMN',
                    zaxis_title='TOTAL VOTES'))

fig1.update_layout(scene = dict(
                    xaxis = dict(
                         backgroundcolor="rgb(200, 200, 230)",
                         gridcolor="white",
                         showbackground=True,
                         zerolinecolor="white",),
                    yaxis = dict(
                        backgroundcolor="rgb(230, 200,230)",
                        gridcolor="white",
                        showbackground=True,
                        zerolinecolor="white"),
                    zaxis = dict(
                        backgroundcolor="rgb(230, 230,200)",
                        gridcolor="white",
                        showbackground=True,
                        zerolinecolor="white",),),
                    width=700,
                    margin=dict(
                    r=10, l=10,
                    b=10, t=10)
                  )

fig1.show()
```
