# US Country Level Presidential Result

### 2008_2016_US_County_Level_Presidential_Results

```python
import pandas as pd 
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv("2008_2016_US_County_Level_Presidential_Results.csv") 
```

```python
df.head()
```

```python
fig1 = px.scatter_3d(df, x='county', y='fips_code', z='total_2008')
fig1.update_traces(marker=dict(size=7,
                              line=dict(width=1,
                                        color='blue')),
                  selector=dict(mode='markers'))

fig2 = px.scatter_3d(df, x='county', y='fips_code', z='total_2012')
fig2.update_traces(marker=dict(size=7,
                              line=dict(width=1,
                                        color='red')),
                  selector=dict(mode='markers'))

fig3 = px.scatter_3d(df, x='county', y='fips_code', z='total_2016')
fig3.update_traces(marker=dict(size=7,
                              line=dict(width=1,
                                        color='yellow')),
                  selector=dict(mode='markers'))

fig3 = go.Figure(data=fig1.data + fig2.data +fig3.data)

fig3.update_layout(scene = dict(
                    xaxis_title='COUNTRY',
                    yaxis_title='FLIPS CODE',
                    zaxis_title='TOTAL VOTES'),
                    width=700,
                    margin=dict(r=30, b=10, l=10, t=10))
fig3.update_layout(title='2008,2012 and 2016 US County Level Presidential Results',width=750, height=750)

fig3.show()
```

### 2020_US_County_Level_Presidential_Results

```python
fig1 = px.scatter_3d(dff, x='county_fips', y='county_name', z='votes_gop')
fig1.update_traces(marker=dict(size=7,
                              line=dict(width=1,
                                        color='blue')),
                  selector=dict(mode='markers'))
fig2 = px.scatter_3d(dff, x='county_fips', y='county_name', z='votes_dem')
fig2.update_traces(marker=dict(size=7,
                              line=dict(width=1,
                                        color='red')),
                  selector=dict(mode='markers'))

fig3 = px.scatter_3d(dff, x='county_fips', y='county_name', z='total_votes')
fig3.update_traces(marker=dict(size=7,
                              line=dict(width=1,
                                        color='yellow')),
                  selector=dict(mode='markers'))

fig4 = go.Figure(data=fig1.data + fig2.data +fig3.data)
fig4.update_layout(scene = dict(
                    xaxis_title='county_fips',
                    yaxis_title='COUNTRY',
                    zaxis_title='TOTAL VOTES'),
                    width=700,
                    margin=dict(r=30, b=10, l=10, t=10))

fig4.update_layout(title='2010 US County Level Presidential Results',width=750, height=750)


fig4.show()
```
