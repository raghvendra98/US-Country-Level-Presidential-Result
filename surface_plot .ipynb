```python
y = ['total_2008', 'total_2012', 'total_2016']
x = dff1['county'].values
fig = go.Figure(data=[go.Surface(z=dff1[y].values, x=y, y=x)])
fig.update_layout(title='2008_2020_US_County_Level_Presidential_Results',width=750, height=750)

fig.update_layout(scene = dict(
                    xaxis_title='Years',
                    yaxis_title='Countries',
                    zaxis_title='TOTAL VOTES'))

fig.show()
```

```python
fig1 = px.scatter_3d(dfff1, x='county', y='fips_code', z='total_2008')
fig1.update_traces(marker=dict(size=7,color='blue',
                              line=dict(width=1,color='blue')),
                  selector=dict(mode='markers'))

fig2 = px.scatter_3d(dfff1, x='county', y='fips_code', z='total_2012')
fig2.update_traces(marker=dict(size=7,color='red',line=dict(width=1,color='red')),
                  selector=dict(mode='markers'))

fig3 = px.scatter_3d(dfff1, x='county', y='fips_code', z='total_2016')
fig3.update_traces(marker=dict(size=7,color='yellow',line=dict(width=1,color='yellow')),
                  selector=dict(mode='markers'))

fig4 = px.scatter_3d(dfff1, x='county', y='fips_code', z='total_2020')
fig4.update_traces(marker=dict(size=7,color='green',line=dict(width=1,color='green')),
                  selector=dict(mode='markers'))

fig = go.Figure(data=fig1.data + fig2.data +fig3.data+fig4.data)
fig.update_layout(scene = dict(
                    xaxis_title='COUNTRY',
                    yaxis_title='county_fips',
                    zaxis_title='TOTAL VOTES'),
                    width=700,
                    margin=dict(r=30, b=10, l=10, t=10))

fig.update_layout(title='2010 US County Level Presidential Results',width=950, height=950)


fig.show()
```
